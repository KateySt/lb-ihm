```
from datasets import load_dataset

poems = load_dataset("biglam/gutenberg-poetry-corpus", split="train")

print("Пример стихотворной строки:", poems[0])

```

```
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```
from transformers import AutoModelForCausalLM, Trainer, TrainingArguments

model = AutoModelForCausalLM.from_pretrained("gpt2")

def tokenize(batch):
    return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=128)

dataset = poems.map(tokenize, batched=True)

training_args = TrainingArguments(
    output_dir="./poet_model",
    evaluation_strategy="epoch",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    logging_dir="./logs"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset
)

trainer.train()
```

```
from transformers import pipeline

generator = pipeline("text-generation", model="./poet_model")

print(generator("Стих о весне:", max_length=80, num_return_sequences=1)[0]["generated_text"])
```

```
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from datasets import load_dataset

poems = load_dataset("biglam/gutenberg-poetry-corpus", split="train")

tokenizer = AutoTokenizer.from_pretrained("gpt2")

tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained("gpt2")

def tokenize(batch):
    return tokenizer(batch["line"], truncation=True, padding="max_length", max_length=128)

dataset = poems.map(tokenize, batched=True, remove_columns=poems.column_names)

```

```
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./poet_model",
    evaluation_strategy="epoch",   # вместо evaluate_during_training
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=3,
    logging_dir="./logs",
    save_strategy="epoch",
    logging_steps=50,
)



trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset
)

trainer.train()
```

```
from transformers import pipeline

generator = pipeline("text-generation", model="./poet_model", tokenizer=tokenizer)

print(generator("О, весна пришла", max_length=50, do_sample=True, temperature=0.8, top_k=50)[0]["generated_text"])
```
